```python
registros_database = base_datos.ejecutar_select("select * from apl1_ingresos_egresos_x_dia")
funciones.mostrar_datos_database(registros_database)
print(type(registros_database))
```

```python
cabecera = ["ID"
            ,   "FECHA_DE_RECAUDACION"
            ,   "PORC_5_SOBRE_TOTAL_REGALIAS"
            ,   "PORC_5_SOBRE_VALOR_CONVENIO_SIN_APORTE"
            ,   "PORC_5_DEL_58_INTERADMINISTRATIVO"
            ,   "VALOR_APORTADO_CONTRATO_POR_FONPACI"
            ,   "POR_42_DEDUCIBLE_DE_INTERADMINSTRATIVOS"]
df = pd.DataFrame(registros_database, columns=cabecera)
print(df.to_markdown(index=False))
```

```python
registros_database = base_datos.ejecutar_select("select ID, FECHA, INGRESOS, EGRESOS from apl1_flujo_caja order by fecha asc")
funciones.mostrar_datos_database(registros_database)
print(type(registros_database))
```

```python
cabecera = ["ID"
            ,   "FECHA"
            ,   "INGRESOS"
            ,   "EGRESOS"]
df = pd.DataFrame(registros_database, columns=cabecera)
print(df.to_markdown(index=False))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['FECHA'], df['INGRESOS'], label='Ingresos', marker='o')
plt.plot(df['FECHA'], df['EGRESOS'], label='Egresos', marker='o')
plt.title('Ingresos y Egresos a lo largo del tiempo')
plt.xlabel('Fecha Mes-Dia-Año')
plt.ylabel('Valor en $')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=90)
formatter = ticker.StrMethodFormatter('${x:,.2f}'.replace('$', 'COP'))
plt.gca().yaxis.set_major_formatter(formatter)
plt.show()
```
